```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint, solve_ivp, solve_bvp
from scipy.optimize import minimize
```

## Problem 1
In a city of approximately three million, suppose five people who have just become infectious have recently
entered the city carrying a certain disease. Assuming each of those individuals has one contact each day that could spread 
the disease, and an average of three days is spent in the infectious state. Here we model how to find the solution of the corresponding
SIR equations using solve_ivp for fifty days, where each time period is half a day. Note percentages 
of each state are used, not the actual number of people in the state. 

At the peak of the infection, how many in the city will still be able to work (assume for simplicity that those who are in the infectious state either cannot go to work or are unproductive, etc.)?

```python
def prob1():
    b = 1   # average number of contacts made each day
    l = 1/3 # fixed proportion of infected indiviuals who will recover each day

    def ode(t, y):
        return np.array([-b*y[0]*y[1], b*y[1]*y[0] - l*y[1], l*y[1]])

    # define the t0 and tf parameters
    t0 = 0
    tf = 50
    times = np.linspace(0,50,100)

    pop = 3000000
    y0 = np.array([(pop-5)/pop, 5/pop, 0])

    sol = solve_ivp(ode, (t0, tf), y0, t_eval=times)

    # plotting
    plt.plot(sol.t, sol.y[0], label='Susceptible')
    plt.plot(sol.t, sol.y[1], label='Infected')
    plt.plot(sol.t, sol.y[2], label='Recovered')
    plt.title('Problem 1')
    plt.xlabel('Time (days)')
    plt.ylabel('Population Proportion')
    plt.legend()
    plt.show()

    return pop - (pop*np.max(sol.y[1]))
```

## Problem 2
On March 11, 2020, New York City had 52 confirmed cases of Covid-19. On that day New York started its 
lock-down measures. The following models what the spread of the virus could have 
been if New York did not implement any measures to curb the spread of the virus over the next 150 days:
    there are approximately 8.399 million people in New York city,
    the average case of Covid-19 lasts for 10 days, 
    and each infected person can spread the virus to 2.5 people.

Results are plotted for each day.

1) At the projected peak, how many concurrent active cases are there?

2) Assuming that about 5% of Covid-19 cases require hospitalization, and using the fact that there are 
about 58,000 hospital beds in NYC, how much over capacity will the hospitals in NYC be at the projected peak?

```python
def prob2():
    '''
    Using the SIR model, plot a graph that models the
    spead of a disease in NYC and answer the questions.
    1. the projected peak concurent cases
    2. capacity deficit at the projected peak

    returns:
        ans1, ans2: answers to questions 1 and 2
    '''
    
    b = 2.5/10  # average number of contacts made each day
    l = 1/10    # fixed proportion of infected indiviuals who will recover each day

    def ode(t, y):
        return np.array([-b*y[0]*y[1], b*y[1]*y[0] - l*y[1], l*y[1]])

    # define the t0 and tf parameters
    t0 = 0
    tf = 150
    times = np.linspace(0,150,151)

    pop = 8399000
    cases = 52
    y0 = np.array([(pop-cases)/pop, cases/pop, 0])

    sol = solve_ivp(ode, (t0, tf), y0, t_eval=times)
    
    # plotting
    plt.plot(sol.t, sol.y[0], label='Susceptible')
    plt.plot(sol.t, sol.y[1], label='Infected')
    plt.plot(sol.t, sol.y[2], label='Recovered')
    plt.title('Problem 2: Modeling Covid-19 in NYC')
    plt.xlabel('Days since Mar 11')
    plt.ylabel('Population Proportion')
    plt.legend()
    plt.show()

    ans1 = np.max(sol.y[1])*pop
    ans2 = (ans1*.05)-58000
    
    return ans1, ans2
    
prob2()
```

# Problem 3

There are 7 billion people in the world. Suppose that influenza originates with 1000 people, and that they are contagious for 10 days before recovering.
Also suppose that on average someone makes one contact every two days that could spread the flu.
Since we can catch a new strain of the flu, suppose that a recovered individual becomes susceptible again with probability f=1/50. To assure a steady population, we let birth-rate μ=.0001.

```python
def prob3():
    '''
    Use the SIRS model to examine the spread of the 
    flu over the course of a year
    '''
    b = .5  # average number of contacts made each day
    l = 1/10    # fixed proportion of infected indiviuals who will recover each day
    m = .0001
    f = 1/50

    def ode(t, y):
        return np.array([f*y[2]+m*(1-y[0])-b*y[0]*y[1], 
                         b*y[1]*y[0] - (l+m)*y[1], 
                         -(f+m)*y[2]+l*y[1]])

    # define the t0 and tf parameters
    t0 = 0
    tf = 365
    times = np.linspace(0,365,366)

    pop = 7000000000
    cases = 1000
    y0 = np.array([(pop-cases)/pop, cases/pop, 0])

    sol = solve_ivp(ode, (t0, tf), y0, t_eval=times)
    
    # plotting
    plt.plot(sol.t, sol.y[0], label='Susceptible')
    plt.plot(sol.t, sol.y[1], label='Infected')
    plt.plot(sol.t, sol.y[2], label='Recovered')
    plt.title('Problem 3: Modeling Influenza Outbreak')
    plt.xlabel('Time (days)')
    plt.ylabel('Population Proportion')
    plt.legend()
    plt.show()
    
prob3()
```

# Problem 4

Fits the PGED model to the Covid-19 data in new_york_cases.npy.
Prints the optimal values of alpha and T_G, and plot your results against 1-S(t).

```python
def prob4():
    '''
    Find the best parameters to fit the PEGD model
    to the data provided in new_york_cases.npy

    Plot the data against 1-S(t)
    '''
    
    data = np.load('new_york_cases.npy')

    def ode(t, y, a, T):
        return np.array([(-a/t)*y[1], ((a/t)-(1-T))*y[1], (1/T)*y[1]])

    # define the t0 and tf parameters
    t0 = 1
    tf = len(data)
    times = np.linspace(1,tf,tf)

    pop = 8399000
    cases = data[0]
    y0 = np.array([(pop-cases)/pop, cases/pop, 0])
    
    # defining optimization function
    def fun(params):
        a, T = params
        sol = solve_ivp(ode, (t0, tf), y0, args=(a, T), t_eval=times)
        out = 1-sol.y[0]
        diff = out - data/pop
        
        return np.linalg.norm(diff)

    # getting optimal solution
    p0 = (1, 1)
    a, T = minimize(fun, p0).x
    sol = solve_ivp(ode, (t0, tf), y0, args=(a, T), t_eval=times)
    
    # plotting
    plt.plot(times, data/pop, label='Data')
    plt.plot(times, 1 - sol.y[0], label='1-S(t)')
    plt.title('Problem 4: Modeling NYC Covid Cases')
    plt.xlabel('Days Since Mar 11')
    plt.ylabel('Population Proportion')
    plt.legend()
    plt.show()

prob4()
```
